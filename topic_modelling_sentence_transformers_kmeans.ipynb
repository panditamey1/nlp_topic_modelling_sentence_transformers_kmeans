```python
import torch
from sentence_transformers import SentenceTransformer, util
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
embeddings = model.encode(data['tweet'])
```

```python
from sklearn.cluster import KMeans
```

```python
kmodel = KMeans(n_clusters = 3)
```

```python
kmodel.fit(embeddings)
```

```python
kmodel.labels_
```

```python
labels = kmodel.labels_

cluster_sent = {}
for index, label in enumerate(labels):
  if label not in cluster_sent:
    cluster_sent[label] = []
  cluster_sent[label].append(index)
```

```python
cluster_sent.keys()
```

```python
cluster1 = data[data.index.isin(cluster_sent[0])]
```

```python
cluster1
```

```python
cluster2 = data[data.index.isin(cluster_sent[1])]
```

```python
cluster2
```

```python
cluster3 = data[data.index.isin(cluster_sent[2])]
```

```python
cluster1.to_csv("cluster_sentence_transformers_1.csv", index = False)
cluster2.to_csv("cluster_sentence_transformers_2.csv", index = False)
cluster3.to_csv("cluster_sentence_transformers_3.csv", index = False)
```
